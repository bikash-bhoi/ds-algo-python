### **Linear Search**
Linear Search is defined as a sequential search algorithm that starts at one end and goes through each element of a list until the desired element is found, otherwise the search continues till the end of the data set.

**O(n)**

![](https://media.geeksforgeeks.org/wp-content/cdn-uploads/Linear-Search.png)

```python
def linearsearch(A, key):
  index = 0
  while index < len(A):
    if A[index] == key:
      return index
    index += 1
  return -1

A = [45,78,88,55,7,96]
found = linearsearch(A, 55)
print(found)
found = linearsearch(A, 100)
print(found)
```

### **Binary Search**
Binary Search is a searching algorithm used in a sorted array by repeatedly dividing the search interval in half. The idea of binary search is to use the information that the array is sorted and reduce the time complexity to O(Log n). 

**O(log n)** : For both Iterative and Recursive methods

![](https://media.geeksforgeeks.org/wp-content/uploads/20220309171621/BinarySearch.png)

**Binary Search Algorithm:** The basic steps to perform Binary Search are:



1.   Sort the array in ascending order.
2.   Sort the array in ascending order.
3. Set the low index to the first element of the array and the high index to the last element.
4. Set the middle index to the average of the low and high indices.
If the element at the middle index is the target element, return the middle index.
5. If the target element is less than the element at the middle index, set the high index to the middle index â€“ 1.
6. If the target element is greater than the element at the middle index, set the low index to the middle index + 1.
7. Repeat steps 3-6 until the element is found or it is clear that the element is not present in the array.

```python
def binary_iterative(A, key):
  l = 0
  r = len(A) - 1
  while l <= r:
    mid = (l + r) // 2
    if key == A[mid]:
      return mid
    elif key < A[mid]:
      r = mid - 1
    else:
      l = mid + 1
  return -1



A = [7,45,55,78,88,96]
found = linearsearch(A, 55)
print(found)
found = linearsearch(A, 100)
print(found)
```

```python
def binary_recursive(A, key, l, r):
  if l > r: return -1
  else:
    mid = (l + r) // 2
    if key == A[mid]:
      return mid
    elif key < A[mid]:
      return binary_recursive(A, key, l, mid-1)
    else:
      return binary_recursive(A, key, mid+1, r)


A = [7,45,55,78,88,96]
found = binary_recursive(A, 55, 0, 5)
print(found)
found = binary_recursive(A, 100, 0, 5)
print(found)
```
