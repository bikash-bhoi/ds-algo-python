```python
class _Node:
  __slots__ = '_element', '_next'
  def __init__(self, element, next):
    self._element = element
    self._next = next
```

```python

class LinkedList:
  def __init__(self):
    self._head = None
    self._tail = None
    self._size = 0

  def __len__(self):
    return self._size
  
  def isempty(self):
    return self._size == 0

  def addfirst(self, e):
    newest = _Node(e, None)
    if self.isempty(): 
      self._tail = newest
    else:
      p = self._head
      newest._next = p
    self._head = newest
    self._size += 1

  def addlast(self, e):
    newest = _Node(e, None)
    if self.isempty():
      self._head = newest
    else:
      self._tail._next = newest
    self._tail = newest
    self._size += 1

  def addany(self, e, position):
    newest = _Node(e, None)
    if position == 0:
      self.addfirst(e)
      return
    i = 0
    p = self._head
    while i < position - 1:
      p = p._next
      i += 1
    newest._next = p._next
    p._next = newest
    self._size += 1

  def removefirst(self):
    if self.isempty():
      print('List is empty')
      return
    e = self._head._element
    self._head = self._head._next
    self._size -= 1
    if self.isempty():
      self._tail = None
    
    print(f"Removed element {e}")
    return e

  def removelast(self):
    if self.isempty():
      print('List is empty')
      return
    i = 1
    p = self._head
    while i < len(self) - 1:
      p = p._next
      i += 1
      print(i, p._element)
    e = p._next._element
    p._next = None
    self._tail = p
    self._size -= 1

    if self.isempty():
      self._head = None
    
    print(f"Removed element {e}")
    return e

  def removeany(self, position):
    if self.isempty():
      print('List is empty')
      return
    p = self._head
    if position == 0:
      self.removefirst()
      return
    i = 0
    p = self._head
    while i < position - 1:
      p = p._next
      i += 1
    e = p._next._element
    p._next = p._next._next
    self._size -= 1

    print(f"Removed element {e}")
    return e

  def search(self, key):
    p = self._head
    index = 0
    while p:
      if key == p._element:
        return index
      else:
        p = p._next
        index += 1
    return -1
      
  def display(self):
    if self.isempty():
      print("Empty")
      return
    p = self._head
    while p:
      print(p._element, end=' --> ')
      p = p._next
    print(f"\t Head : {self._head._element}, Tail : {self._tail._element}")
```

```python
## Addition 
L = LinkedList()
L.addlast(5)
L.addlast(7)
L.addlast(12)
L.display()
print(f"Size : {len(L)}")
L.addlast(8)
L.addlast(3)
L.display()
idx = L.search(89)
print(f"search index : {idx}")
L.addfirst(1)
L.addfirst(4)
L.display()
L.addany(99, 2)
L.display()
L.addany(28, 3)
L.display()
```

```python
## Deletion  
L = LinkedList()
L.addlast(5)
L.addlast(7)
L.addlast(8)
L.addlast(3)
L.addlast(17)
L.addlast(18)
L.addlast(13)
L.addany(99, 2)
L.display()
L.removefirst()
L.display()
L.removelast()
L.display()

L.removeany(3)
L.display()
L.removeany(0)
L.display()

L.removeany(3)
L.display()
```
